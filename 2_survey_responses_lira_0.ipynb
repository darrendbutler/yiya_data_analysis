```python
# Keep only survey repsonses
lira_responses_survey = lira_responses.loc[lira_responses["key"].isin(profile_survey_keys)]

#Only keep the last responses to a question (Remove duplicates)
lira_responses_survey =  lira_responses_survey.drop_duplicates(subset=["user_id", "key"], keep="last")
lira_responses_survey.sample(5, random_state=5)
```

```python
#Strip excess info from string
key_pattern = r"(profile/.+|(?<=airscience-2022a/course/).+)"
question_column = lira_responses_survey["key"].str.extract(key_pattern)
lira_responses_survey["question_name"] = question_column
```

```python
lira_responses_survey = lira_responses_survey.pivot(index="user_id",columns="question_name",values="value")
lira_responses_survey
```

```python
# Visualize Survey Results
def visualize_question(df, question):
    #counts = df[question].value_counts(normalize=True)
    ax = df[question].value_counts(normalize=True).plot.bar()
    plt.title(question + " (N = {})".format(len(lira_responses_survey)))
    plt.ylabel("Frequency (%)")
    plt.xlabel("Response Options")

    fig = ax.get_figure()
    folder = "/Users/ddbutler/repos_new/yiya_data_analysis/visualizations/lira_learning_correlations/"
    file_name = question.replace("/", "_")
    plt.xticks(rotation = 25)
    plt.savefig(folder + file_name, bbox_inches="tight")

    #fig.savefig("/Users/ddbutler/repos_new/yiya_data_analysis/visualizations/lira_learning_correlations/" + '{}'.format(question))
    plt.close()

for question in lira_responses_survey.columns:
    visualize_question(lira_responses_survey, question=question)

#visualize_question(lira_responses_survey, "profile/age")
```
